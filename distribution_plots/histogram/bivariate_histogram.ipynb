```python
# Bivariate histogram

# Bivariate histogram bins the data within rectangles(box) that tile the plot and then shows the count of observation within each rectangle with the fill color
```

```python
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
tip = sns.load_dataset("tips")
print(tip)
```

```python
# Here we are creating bivariate histogram of tips and total bills

plt.rcParams["font.family"] = "Cascadia Code"
plt.rcParams["font.size"] = 14
# sns.displot(kind="hist", data=tip, x="total_bill", y="tip")

sns.displot(kind="hist", data=tip, x="total_bill")
```

```python
sns.displot(kind="hist", data=tip, y="tip")
```

```python
# Here we are creating bivariate histogram of tips and total bills using x and y parameters

sns.displot(kind="hist", data=tip, x="total_bill", y="tip")

"""
Here the bivariate plot gave the combination of occurence of pair of totalbill and tip.
Where the darker color reprsent most occurance and the lighter one reprsent the least occurance. 

The value of lets say 15 in total bill have maximum number of tips which is almost 2.5 dollars.
The most lighter rectangular box at 50,10 represent least number of people lies in the range where the total bills is 50 and gave tip of 10
"""

```
