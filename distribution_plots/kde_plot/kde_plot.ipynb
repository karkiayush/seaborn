```python
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
# kde plot means kernel density estimation plot

# Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate
```

```python
tips = sns.load_dataset("tips")
print(tips)
```

```python
plt.rcParams["font.family"] = "Monolisa Trial"
plt.rcParams["font.size"] = 14
sns.displot(kind="kde", data=tips, x="total_bill")


# kde plot gives the density of occurance of particular parameters of x axis.

# Here in the below plot, we can say the probability of total bill to be 20 is around 0.05 => 5%
```
