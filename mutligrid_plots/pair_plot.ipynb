```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
plt.rcParams["font.family"] = "Mooli"
plt.rcParams["font.size"] = 15
# Pair plot: It plots the relationship pair wise
iris = sns.load_dataset("iris")

# From the numerical columns of the data frame, it picks up a pair and plots scatter plot
# For example: sepal_length vs sepal width, sepal width vs petal length, petal_length vs petal width, petal width vs sepal length and so on.
sns.pairplot(iris,hue="species")

# same parameter vs same parameter yield histogram
```
