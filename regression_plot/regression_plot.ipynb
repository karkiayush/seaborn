```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
# Plotting regression plot
plt.rcParams["font.family"] = "Monolisa Trial"
plt.rcParams["font.size"] = 15

sns.regplot(data=tips, x="total_bill", y="tip")

# Here the shade around the linear graph is the error of the line which can occur.

# In general the equation lies with inthe area of shades
```

### Regression Plots

- regplot
- lmplot

In the simplest invocation, both functions draw a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression.


```python
# Figure level function
sns.lmplot(data=tips, x="total_bill", y="tip", hue="sex")
```

```python
# <<------ residual plot------>>
sns.residplot(data=tips, x="total_bill", y="tip")

# Here the residual plot gives the info about the error rate.
# It firstly plots the scatter plot and internally plots the regression plot as well and at last plotting the middle line, it shows the how far the point lies from the regression plot line

# Here we can say, the upper point says, that the model provide low value than it should be &  the lower points tell that our model provide higher value than it should be .
```
