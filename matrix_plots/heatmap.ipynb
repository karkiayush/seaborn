```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Heatmap

# it plots the rectangular data as a color-encoded matrix, where each color will be representing the value or any meaningful information
```

```python
country_data = px.data.gapminder()
country_data
```

```python
# constructing out new filtered table where based on the year, we showed the life expectency of every country.
temp_country_df = country_data.pivot(index="country", columns="year", values="lifeExp")

temp_country_df
```

```python
# Now the above grid is represented in heatmap as:
plt.rcParams["font.family"] = "Monolisa Trial"
plt.rcParams["font.size"] = 14

plt.figure(figsize=(17, 10))

sns.heatmap(temp_country_df)
```

```python
nepal_df = country_data[country_data["country"] == "Nepal"]

nepal_df
```

```python
nepal_grid_df = nepal_df.pivot(index="country", columns="year", values="lifeExp")
nepal_grid_df
```

```python
plt.figure(figsize=(17, 10))
sns.heatmap(data=nepal_grid_df, annot=True)
```

```python
# Heatmap of life expectency of Europe

eur_df = country_data[country_data["continent"] == "Europe"]

eur_df
```

```python
eur_df_grid = eur_df.pivot(index="country", columns="year", values="lifeExp")

eur_df_grid
```

```python
plt.figure(figsize=(20, 15))
sns.heatmap(data=eur_df_grid, annot=True, linewidths=0.5, cmap="viridis")
```
