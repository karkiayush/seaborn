```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import numpy as np
```

```python
# categorical estimation plot which consists of bar plot, point plot and count plot are used to calculate central tendency in general.

# For the calculation of some value like mean, median or anything
```

```python
tips = sns.load_dataset("tips")
```

```python
# <---------Bar plot-------->

# Bar plot helps to calculate the aggregrate value of one categorical coloumns like sex, time with the numerical coloumns like total bills, tips etc.

# By default it will calculate the mean, but we can put median, max, standard deviation anything as a estimator
```

```python
plt.rcParams["font.family"] = "Monolisa Trial"
plt.rcParams["font.size"] = 15
plt.figure(figsize=(14, 7))

sns.barplot(data=tips, x="sex", y="total_bill", hue="sex")

# When there are multiple observation in each category like sex: male, female, female, male and so on. Here boostrapping(statistical technique) is used to compute a confidence interval around the estimate which is plotted using error bars.

# Hence the black marking is called error bars which can be disabled as well.
```

```python
# making the standard deviation of the total bill as the estimator.
sns.barplot(data=tips, x="sex", y="total_bill", hue="sex", estimator=np.std)
```

```python
# <-------Point Plots-------->
# point plot joins the point of error bars to depict the difference in estimator values of two category.
```

```python
sns.pointplot(data=tips, x="sex", y="total_bill", hue="smoker")
```

```python
sns.barplot(data=tips, x="sex", y="total_bill", hue="smoker")
```

```python
# <---------count plot---------->
# count plot is a special case for the bar plot is when you want to show the number of observations in each category rather than computing a statistic for a second variable. This is similar to a histogram over a categorical, rather than quantitative, variable
```

```python
sns.countplot(data=tips, x="sex", hue="sex")
```

```python
# It gives the count of every male and female arrving in each day: thursday, friday, saturday and sunday
sns.catplot(kind="count", data=tips, x="sex", hue="day")
```
