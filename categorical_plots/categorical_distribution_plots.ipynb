```python
# Categorical distribution plots is used for univariate analysis as we know that distribution plot is used for analysing single variable. On the other hand, the categorical scatter plot is used for bivariate analysis.
```

```python
# <----Box Plot------>
# A boxplot is a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile [Q1], median, third quartile [Q3] and “maximum”). It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed(jhukab)
```

```python
# Here 5 number summary have following significance:

# 1. Q1(25th percentile): That data where 25 % data is less from total data

# 2. Q2(75th percentile): That data where 75 % data is less from total data

# 3. Median(50th percentile): That data where 50 % data is less from total data

# IQR(Interquartile Range): Difference of Q3-Q1

# 4.Minimum: It is not the actual minimum of our data. It is the minimum, which is obtained by calculation using that formula. Some value might be lesser than the minimum value, which are known to be outliers

# 5.Maximum: It is not the actual maximum of our data. It is the max value, which is obtained by calculation using that formula. Some value might be larger than the maximum value, which are known to be outliers

# Though box plot is the distribution plot, that gives the distribution of one cols, Seaborn gives us that ability to plot the box plot(categorical distribution plot) with another categorical coloumns.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
tip = sns.load_dataset("tips")
tip
```

```python
plt.rcParams["font.family"] = "Monolisa Trial"
plt.rcParams["font.size"] = 15
plt.figure(figsize=(14, 7))

sns.boxplot(data=tip, y="total_bill", x="day", hue="sex")
```

```python
sns.boxplot(data=tip, y="total_bill")
```

# Violin Plot

#### violin plot=Box plot+kde plot


```python
plt.figure(figsize=(14, 7))
sns.violinplot(data=tip, x="day", y="total_bill", hue="sex")
```

```python
plt.figure(figsize=(14, 7))
sns.violinplot(data=tip, x="day", y="total_bill", hue="sex", split=True)
```
