### Figure level function -> `catplot`


```python
# importing some datasets
tips = sns.load_dataset("tips")

iris = sns.load_dataset("iris")
tips["time"]
```

```python
# Categorical plots are those type of plots which are used for categorical data

# categorical scatter plot, categorical distribution plot and categorical estimate plot are 3 types of categorical plots.

""" As we know generally, we use Scatter plots for visualizing numerical data. Like for any instant of x, what is the value that y have at that instant of x. 


But if we want, then we can use the Scatter plot for categorical data as well, but we need to make sure that one coloumn is categorical and one cols is numerical. 

Categorical scatter plot:
1. strip plot
2. swarm plot
"""
```

```python
# strip plot: it pick one categorical col and one numerical col, and plots the scatter plot.

plt.rcParams["font.family"] = "Monolisa Trial"
plt.rcParams["font.size"] = 15

plt.figure(figsize=(15, 7))
sns.stripplot(data=tips, x="day", y="total_bill", hue="time")
```

```python
# usually the data points are seperated by some margins to improve the readability. Since seaborn add some additional noise, known to be gitter that seprates the data points by some distance

# It can be removed by using the jitter attribute = False

plt.figure(figsize=(15, 7))
sns.stripplot(data=tips, x="day", y="total_bill", hue="time", jitter=False)


# Therefore the categorical scatter plots gives the information about where the numerical value of every category lies.
```

```python
# stripplot was the axes level function

# catplot() is the figure level function
plt.rcParams["font.family"] = "Monolisa Trial"
sns.catplot(kind="strip", data=tips, x="day", y="total_bill", hue="time", jitter=0.1)
```

```python
# Swarm plot

"""
Swarm plots are similar to strip plots but with one key difference: they adjust the position of points to avoid overlap.

Points are spread out along the categorical axis, ensuring that no two points overlap vertically.
"""
plt.rcParams["font.size"] = 15
sns.catplot(kind="swarm", data=tips, x="day", y="total_bill", hue="day")
```
