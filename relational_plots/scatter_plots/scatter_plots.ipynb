```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# plotly.express is imported as it contains some data sets which we gonna use directly
```

```python
# tips is built in data set of seaborn
tips = sns.load_dataset("tips")
print(tips)
```

```python
# Here we are plotting scatter plots taking total bills and tip in x and y axis
plt.rcParams["font.family"] = "Monolisa Trial"
plt.rcParams["font.size"] = 12
sns.scatterplot(data=tips, x="total_bill", y="tip")

# This scatterplot is an axes level function
```

```python
# For the figure level scatter plot function
sns.relplot(data=tips, x="total_bill", y="tip", kind="scatter")

# relplot means relational plot

# Only the change is the width and height of the figure is changed.

# relplot() is figure level function which is generally a square shaped plot

# Seaborn recommend to use figure level function more due the reason that with the help of figure level function, we can replicate many axes level function.

# For eg: line plot can be plotted using linepal_dfal_dfal_dflot standard axes level function. Instead we can also use this relplot where we provide kind = "line" in the kind attribute
```

```python
plt.rcParams["font.family"] = "Monolisa Trial"
plt.rcParams["font.size"] = 12
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="sex",
    style="time",
    size="size",
    # depending upon the value of size coloumn, the marker size will be increased or decreased
    # Here hue parameter helps to know the gender of guest by distinguishing them with different color
    # style gives a distinction of lunch and dinner time
)
```

```python
# facet plot
# facet plot is the type of plot where based on some category, we can plot multiple plots of same data

# facet plot is only possible to draw with figure level functions i.e. it only works in relplot function but not with line plot and scatter plot function since we are doing the manipulation in the whole figure level hence we need the control of whole figure level and we know that axes plot only have the control of its axes but not whole figure
plt.rcParams["font.size"] = 14
sns.relplot(data=tips, x="total_bill", y="tip", kind="scatter", col="sex", size="size")
```

```python
plt.rcParams["font.size"] = 15
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    kind="scatter",
    row="smoker",
    col="sex",
    size="size",
)
```
