```python
# Line plot is upgraded version of scatter plot where the marker are connected with the help of lines.

# Hence line plot is used to represent continuity in our graph.

# Usually, if we have time value in the x axis, we use line plot since line plots demonstrate the time series data in better manner
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# line plot
gap = px.data.gapminder()
gap
```

```python
# Plotting the variation in life expentency of any specific nation from 1952-2007 A.D.

nepal_df = gap[gap["country"] == "Nepal"]
nepal_df
# plotting life expectency w.r.t year
```

```python
plt.rcParams["font.family"] = "Monolisa Trial"
plt.rcParams["font.size"] = 12
sns.lineplot(data=nepal_df, x="year", y="lifeExp")
```

```python
plt.rcParams["font.family"] = "Monolisa Trial"
plt.rcParams["font.size"] = 12

sns.relplot(data=nepal_df, x="year", y="lifeExp", kind="line")
```

```python
# Lets plot the life expectency of nepal, india and pakistan side by side and analyse whose life expectency is better and worst in particular year

mask = gap["country"].isin(["Nepal", "Zimbabwe", "Germany"])
nep_usa_zimbawe_df = gap[mask]
nep_usa_zimbawe_df
```

```python
plt.rcParams["font.family"] = "Monolisa Trial"
plt.rcParams["font.size"] = 15

sns.relplot(
    data=nep_usa_zimbawe_df,
    kind="line",
    x="year",
    y="lifeExp",
    hue="country",
    style="continent",
    size="iso_num",
    # depending upon the value of size coloumn, the marker size will be increased or decreased
    # Here hue parameter helps to know the country life expectency by distinguishing them with different color
    # style gives a distinction of lunch and dinner time
)
```
